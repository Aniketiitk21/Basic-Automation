```python
import cv2
import torch
from ultralytics import YOLO
from google.colab.patches import cv2_imshow  # Use this for displaying images in Colab

# Load the trained YOLO model
model = YOLO("/content/drive/MyDrive/Dailton Daraset/runs/detect/train/weights/best.pt")  # Load custom model

# Load the image
image_path = "/content/drive/MyDrive/Dailton Daraset/YOLO-Military-10/test/images/100_jpg.rf.59a5d852298d6ff8e8ee1214ba11f173.jpg"
image = cv2.imread(image_path)

# Run inference
results = model(image_path)

# Get class names
class_names = model.names

# Draw bounding boxes on the image
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Get bounding box coordinates
        conf = box.conf[0].item()  # Confidence score
        cls = int(box.cls[0].item())  # Class index

        label = f"{class_names[cls]} {conf:.2f}"
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Draw bounding box
        cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)  # Add label

# Show the image with bounding boxes in Colab
cv2_imshow(image)
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load the trained YOLOv8 model
model_path = "/content/bestmode.pt"
model = YOLO(model_path).model  # Extract model layers

# Store sparsity of each layer
layer_sparsity = {}

for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        weight = module.weight.data.cpu().numpy()
        sparsity = np.mean(np.abs(weight) < 1e-3)  # Count near-zero weights
        layer_sparsity[name] = sparsity

# Sort layers by sparsity
sorted_layers = sorted(layer_sparsity.items(), key=lambda x: x[1], reverse=True)

# Display sparsity results
for layer, sparsity in sorted_layers:
    print(f"Layer: {layer} - Sparsity: {sparsity:.2f}")

# Plot sparsity distribution
plt.figure(figsize=(20, 10))
plt.barh([x[0] for x in sorted_layers], [x[1] for x in sorted_layers])
plt.xlabel("Sparsity Ratio (Higher means more prunable)")
plt.ylabel("Layer Name")
plt.title("Sparsity Analysis of YOLOv8 Layers")
plt.show()
```

```python
import torch
from torch.nn.utils import prune
from ultralytics import YOLO

# Load YOLO model
model_path = "/content/bestmode.pt"
model = YOLO(model_path).model  # Extract YOLO layers

# List of layers to prune (Reducing pruning amount & skipping critical layers)
layers_to_prune = [
    "model.22.cv2.2.1.conv",
    "model.22.cv2.2.0.conv",
    "model.22.cv2.2.2",
    "model.22.cv2.1.1.conv",
    "model.22.cv2.1.0.conv",
    "model.22.cv2.0.1.conv",
]

# Apply pruning (Reduce from 15% → 10%)
for layer_name in layers_to_prune:
    module = dict(model.named_modules()).get(layer_name)
    if isinstance(module, torch.nn.Conv2d):
        prune.l1_unstructured(module, name="weight", amount=0.10)  # Lower pruning
        prune.remove(module, "weight")  # Make pruning permanent
        print(f"✅ Pruned layer: {layer_name}")

# 🔥 Recalibrate BatchNorm (More careful)
for name, module in model.named_modules():
    if isinstance(module, torch.nn.BatchNorm2d):
        module.momentum = 0.01  # Reduce momentum to preserve statistics instead of erasing them

# Save pruned model correctly
pruned_model_path = "/content/bestmode_pruned.pt"
torch.save(model.state_dict(), pruned_model_path)
print(f"✅ Pruned model weights saved at {pruned_model_path}")
```

```python
# Load the pruned model
pruned_model = YOLO("/content/bestmode.pt")  # Load structure from original model
pruned_model.model.load_state_dict(torch.load("/content/bestmode_pruned.pt"))

# Fine-tune pruned model (3-5 epochs)
pruned_model.train(data="/content/drive/MyDrive/Dailton Daraset/YOLO-Military-10/data.yaml", epochs=5)

# Save fine-tuned model
pruned_model.save("/content/bestmode_pruned_finetuned.pt")
print("✅ Fine-tuned pruned model saved successfully!")
```

```python
# Load original and fine-tuned pruned models
original_model = YOLO("/content/bestmode.pt")
pruned_model = YOLO("/content/bestmode_pruned_finetuned.pt")

# Validate both models
print("Validating Original Model:")
original_model.val(data="/content/drive/MyDrive/Dailton Daraset/YOLO-Military-10/data.yaml")

print("Validating Fine-Tuned Pruned Model:")
pruned_model.val(data="/content/drive/MyDrive/Dailton Daraset/YOLO-Military-10/data.yaml")
```
